```python
root_dir = '/home/KDT-admin/data/images'
output_path = '/home/KDT-admin/data/images/cropped'
train_img_list = sorted(os.listdir(os.path.join(root_dir, 'train')))
val_img_list = sorted(os.listdir(os.path.join(root_dir, 'val')))
```

```python
# 

input = os.path.join(root_dir, 'train', train_img_list[0])
print(input)
```

```python
for v in train_img_list:
    input = os.path.join(root_dir, 'train', v)
    img = cv2.imread(input)
    face = DeepFace.detectFace(img_path=img,
                        detector_backend='mediapipe',enforce_detection=False)
    if face is None:
        continue
    plt.imsave(os.path.join(output_path, 'train', v), cv2.cvtColor(face, cv2.COLOR_BGR2RGB))
    #cv2.imwrite(os.path.join(output_path, 'train', v), cv2.cvtColor(face, cv2.COLOR_BGR2RGB))
```

```python
print(len(sorted(os.listdir(os.path.join(root_dir, 'cropped','train')))))
print(len(sorted(os.listdir(os.path.join(root_dir, 'cropped','val')))))
```

```python
for v in val_img_list:
    input = os.path.join(root_dir, 'val', v)
    img = cv2.imread(input)
    face = DeepFace.detectFace(img_path=img,
                        detector_backend='mediapipe',enforce_detection=False)
    if face is None:
        continue
    plt.imsave(os.path.join(output_path, 'val', v), cv2.cvtColor(face, cv2.COLOR_BGR2RGB))
```
