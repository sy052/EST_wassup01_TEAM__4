```python
import soundfile

from espnet2.bin.tts_inference import Text2Speech

text2speech = Text2Speech.from_pretrained('mio/amadeus')

print('TTS model loaded!')
```

```python
from IPython.display import HTML, Audio
from IPython.display import display

sound_path = '/home/KDT-admin/work/bonghoon/EST_wassup01_TEAM__4/tools/Wav2Lip_copy/contents/out.wav'
text = '  ' #@param {type:"string"}
input_audio_path = sound_path #@param {type:"string"}

speech = text2speech(text)['wav']
soundfile.write(sound_path, speech.numpy(), text2speech.fs, 'PCM_16')

display(Audio(input_audio_path))
```
